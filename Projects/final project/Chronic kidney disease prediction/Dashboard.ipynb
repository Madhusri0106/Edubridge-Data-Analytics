```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Kidney_disease.csv")
data
```

```python
pd.pandas.set_option('display.max_columns', None)
data
```

```python
X = data[['sg', 'htn', 'hemo', 'dm', 'al', 'appet', 'rc', 'pc','pe','sc']]
data['htn'] = data['htn'].replace(to_replace = {'yes' : 1, 'no' : 0})
data['dm'] = data['dm'].replace(to_replace = {'yes' : 1, 'no' : 0})
data['appet'] = data['appet'].replace(to_replace={'good':1,'poor':0})
data['pe'] = data['pe'].replace(to_replace = {'yes' : 1, 'no' : 0})
data['pc'] = data['pc'].replace(to_replace = {'normal' : 0, 'abnormal' : 1})
X.head()
```
